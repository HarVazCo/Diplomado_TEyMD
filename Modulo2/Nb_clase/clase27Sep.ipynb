# Distribuciones muestrales

## Distribución Ji-Cuadrada (Chi-Cuadrada)

 - Si $X\sim N(0,1)$
 - $X^2\sim \chi^2(1)$

 y si 

 $$\sum_{i=1}^{n} X_i^2 \sim \chi^2(n) $$
 donde $X_i$ son v.a.i con distribución normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Damos un rango de valores para las gráficas
x_chi = np.linspace(0,20,500)
```

```python
# Grafica de la distribucion chi-cuadrada con dif grados de libertad
plt.figure(figsize=(10,6))

df_chi_list = [1,3,6,10] # Grados de libertad

for df in df_chi_list:
    plt.plot(x_chi,chi2.pdf(x_chi,df),label=f'Chi-cuadrado (df={df})')

plt.title('Distribución Chi-cuadrado')
plt.xlabel('x') # Etiqueta eje x
plt.ylabel('Densidad de probabilidad') # Etiqueta eje y
plt.legend()
plt.grid()
plt.show()
```

## Distribución $t$ de Student

Sean 
 - $X\sim N(0,1)$
 - $Y\sim \chi^2(n-1)$

 Entonces
 $$ T = \frac{X\cdot\sqrt{n}}{\sqrt{\frac{Y}{n-1}}} \sim t(n-1) $$

 Podemos decir que

 $$\lim_{n\to\infty} t(n) = N(0,1) $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t,norm

# Damos un rango de valores para las graficas
x = np.linspace(-5,5,500)
```

```python
# Grados de libertad
dfs = [1,3,10,30,50]

# Colores
colo = ['blue','orange','green','red','purple']

# Grafica
plt.figure(figsize=(10,6))

# Graficamos las distribuciones t de Student
for df,color in zip(dfs,colo):
    plt.plot(x,t.pdf(x,df),label=f't (df = {df})',c=color)

# Graficamos la distribucion normal
plt.plot(x,norm.pdf(x),'k--',label='Normal estándar')

plt.title('Comparación entre la distribución t y la normal estándar')
plt.xlabel('x') # Etiqueta eje x
plt.ylabel('Densidad de probabilidad') # Etiqueta eje y
plt.legend()
plt.grid()
plt.show()
```

## Distribución F de Fisher

$$ F = \frac{\chi^2_{(m-1)}/(m-1)}{\chi^2_{(n-1)}/(n-1)} \sim F(m-1,n-1)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Definimos un rango de valores
x = np.linspace(0,5,500)
```

```python
# Lista grados de libertad
df_lista = [(1,2),(2,5),(5,10),(10,20)] # (df1,df2)

plt.figure(figsize=(10,6))

# Graficamos las distribuciones f
for df1,df2 in df_lista:
    plt.plot(x,f.pdf(x,df1,df2),label=f'F (df1={df1}, df2={df2})')

plt.title('Distribución F de Fisher')
plt.xlabel('x') # Etiqueta eje x
plt.ylabel('Densidad de probabilidad') # Etiqueta eje y
plt.legend()
plt.grid()
plt.show()
```
